```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# load dataset

data = load_iris()

x = data.data

# trianing

kmeans = KMeans(n_clusters= 3 , random_state=42)

kmeans.fit(x)

y_pred = kmeans.labels_

print("cluster labels :",y_pred)

plt.figure(figsize=(11,8))

plt.scatter(x[:,0],x[:,1],c = y_pred)

plt.xlabel("feature 1")
plt.ylabel("feature 2")

plt.legend(

   [ "cluster 0 = Iris - Setrosa",
    "cluster 2 = Iris -Vesicolar",
    "cluster 3 = Iris - Virdicinia"]
)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


smaple = np.array([[4.3,3.4,1.5,0.2]])

sample_cluster = kmeans.predict(smaple)

print("smaple belongs to the cluster",sample_cluster)

plt.scatter(x[:,0],x[:,1],c= y_pred)
plt.scatter(smaple[:,0],smaple[:,1],s=200,marker='*',color = "red",label = 'sample')


plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.legend()
plt.show()
```
