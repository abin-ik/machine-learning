```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
load_iris()
```

```python
data = load_iris()

x = data.data

y = data.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

model = KNeighborsClassifier(n_neighbors=5)

model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print("accuracy_score",accuracy_score(y_test,y_pred))
```

```python
new = [[6.2, 3.4, 5.4, 2.3]]

new_data = model.predict(new)

print(f"the predicted data = {data.target_names[new_data]}")
```

```python
plt.scatter(x_test[:,2],x_test[:,3] , c= y_test , label = "actual data" ,marker='o' , cmap='viridis')


plt.scatter(x_test[:,2],x_test[:,3] , c= y_pred , label = "predicted data", marker= 's',cmap='viridis' )

plt.xlabel("petal_length")

plt.ylabel("petal_width")

plt.colorbar(label = "class (0 = setosa , 1 = versicolor , 2 = virginica)")

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scalar = StandardScaler()

x_scalar = scalar.fit_transform(x)
```

```python
pca = PCA(n_components=2)

x_pca = pca.fit_transform(x_scalar)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
for i in range(len(set(y))): #set(y) removes duplicates
    plt.scatter(x_pca[y == i,0], x_pca[y == i,1], label=f'class{i}',edgecolor='k',s=50)
plt.xlabel("PCA1")
plt.ylabel("PC2")
plt.title("PCA - Iris Dataset")
plt.legend()
plt.show()
```
