```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
data = {
    "TV_ads" : [10,12,15,20,22,25],
    "radio_ads" :[5,8,10,12,16,18],
    "sales" : [120,150,170,200,230,260]
}
```

```python
df = pd.DataFrame(data)

x = df[["TV_ads"]]
y = df["sales"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
model = LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print("root_mean_squared_error",root_mean_squared_error(y_test,y_pred))
print("r2_score",r2_score(y_test,y_pred))


```

```python
plt.scatter(x,y,color = "red",label = "actual_data")
plt.plot(x,model.predict(x),color = "blue",label = "regrssion line")
plt.xlabel("tv_ads")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
tv = 18
new_data = pd.DataFrame({"TV_ads":[tv]})
predicted_score = model.predict(new_data)
print(f"tv_ads {tv} sales is {predicted_score[0]:2f}")
```

```python

x1 = df[["radio_ads"]]
y1 = df["sales"]


x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=0.2, random_state=42)

model1 = LinearRegression()
model1.fit(x_train1, y_train1)

y_pred1 = model1.predict(x_test1)

print("root_mean_squared_error", root_mean_squared_error(y_test1, y_pred1))
print("r2_score", r2_score(y_test1, y_pred1))

```

```python
plt.scatter(x1,y1,color= "green",label = "actual data")
plt.plot(x1,model1.predict(x1),color = "yellow",label = "regression line")
plt.xlabel("radio_ads")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
radio = 11
new_data1 = pd.DataFrame({"radio_ads":[radio]})
predicted_score1 = model1.predict(new_data1)
print(f"radio ads {radio} sales is {predicted_score}")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# --- Subplot 1 : Radio Ads ---
ax[0].scatter(x1, y1, color="green", label="actual data")
ax[0].plot(x1, model1.predict(x1), color="yellow", label="regression line")
ax[0].set_xlabel("radio_ads")
ax[0].set_ylabel("sales")
ax[0].set_title("Radio Ads vs Sales")
ax[0].legend()

# --- Subplot 2 : TV Ads ---
ax[1].scatter(x, y, color="red", label="actual data")
ax[1].plot(x, model.predict(x), color="blue", label="regression line")
ax[1].set_xlabel("tv_ads")
ax[1].set_ylabel("sales")
ax[1].set_title("TV Ads vs Sales")
ax[1].legend()

plt.tight_layout()
plt.show()

```
