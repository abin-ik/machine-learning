```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car',as_frame=True).frame
data
```

```python
columns_to_encode = ['lug_boot','safety']
encoder = OrdinalEncoder(
    categories=[
        ['small','med','big'],
        ['low','med','high']
    ]
)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
data
```

```python
data  = fetch_openml('adult',as_frame=True).frame
data.occupation.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
encoded_values = encoder.fit_transform(data[['occupation']])
new_colums= encoder.get_feature_names_out(['occupation'])
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,accuracy_score,precision_score,recall_score,f1_score


x,y = load_breast_cancer(return_X_y=True)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.fit_transform(x_test)

model= KNeighborsClassifier()
model.fit(x_train_scaled,y_train)
y_pred = model.predict(x_test_scaled)

print(r2_score(y_test,y_pred))
```

```python
model1 = KNeighborsClassifier()
model1.fit(x_train,y_train)
y_pred1 = model1.predict(x_test)

print(r2_score(y_test,y_pred1))
```

```python
from sklearn.pipeline import Pipeline

from sklearn.decomposition import PCA

pipe = Pipeline([
    ('scale',StandardScaler()),
    ('pca',PCA(n_components=10)),
    ('KNeighborsClassifier',KNeighborsClassifier())
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)

print(r2_score(y_test,y_pred))
```
