```python
import tensorflow as tf
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

model = tf.keras.Sequential([
    tf.keras.Input(shape =(1,)),
    tf.keras.layers.Dense(1)
])
```

```python
#compile the model
model.compile(optimizer='sgd',loss='mean_squared_error')

#train the model

model.fit(x,y,epochs=200,verbose=1)     #epochs = views the dataset n times . verbose = if you want to see the training progress 

print(model.predict(np.array([7])))

```

```python
# 1️⃣ Customer Age vs Purchase Probability (Sigmoid)
# Training Data:
# Customer Age: [18, 25, 35, 45, 60]
# Purchase Probability: [0.1, 0.3, 0.5, 0.7, 0.9]
# Question: Build a neural network with 1 hidden layer using sigmoid activation to predict the purchase probability for a 50-year-old customer.
```

```python
customer_age = np.array([18,25,35,45,60], dtype=np.float32)
purchase = np.array([0.1,0.3,0.5,0.7,0.9], dtype=np.float32)



model6 = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model6.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
               loss='mse')

model6.fit(customer_age, purchase, epochs=200, verbose=0)



print(model6.predict(np.array([50])))
```

```python
# 4️⃣ Daily Steps vs Calories Burned (Tanh)
# Training Data:
# Steps: [1000, 3000, 5000, 7000, 10000]
# Calories Burned: [50, 150, 250, 350, 500]
# Question: Build a neural network with Tanh activation to predict calories burned for 8000 steps.
```

```python
daily_steps = np.array([1000,3000,5000,7000,10000])
calories = np.array([50,150,250,350,500])

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(16),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.fit(daily_steps, calories, epochs=200, verbose=0)

print(model.predict(np.array([8000])))
```

```python
# 5️⃣ Advertisement Impressions vs Click Probability (Sigmoid)
# Training Data:
# Impressions (thousands): [10, 20, 30, 40, 50]
# Click Probability: [0.01, 0.05, 0.1, 0.15, 0.2]
# Question: Build a neural network with sigmoid activation in the output layer to predict click probability for 60k impressions.
```

```python
imprression = np.array([10,20,30,40,50])
prob = np.array([0.01, 0.05, 0.1, 0.15, 0.2])

model8 = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(4,activation="sigmoid"),
    tf.keras.layers.Dense(1),
])

model8.compile(optimizer="sgd",loss= "mse")
model8.fit(imprression,prob,epochs=200,verbose=0)

print(model8.predict(np.array([60])))
```
