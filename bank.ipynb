a. Design a BankAccount class that allows users to:
      • Deposit money
      • Withdraw money 
      • Check account balance

   b. The account balance must not go below 0.

   c. The class must include a method to calculate yearly interest.

   d. Create a SavingsAccount class that inherits from BankAccount and applies a higher interest rate.

   e. Constraints:
      • Interest rate for BankAccount: 2% per annum
      • Interest rate for SavingsAccount: 3% per annum
      • Negative balance is not allowed
